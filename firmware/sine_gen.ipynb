```python
import numpy as np
import matplotlib.pyplot as plt
import os

file_name = 'sine.mem'

print(os.path.abspath(file_name))
```

```python
t = np.arange(0, 256, 1).astype(int)
s = (128 + np.sin(np.pi * t / 32)*127).astype(int)
print(s.min())
print(s.max())
plt.plot(t,s)
```

```python
with open(file_name, 'w') as sine_file:
    for i in s:
        hx = ('00' + hex(i)[2:])[-2:]
        print(hx, end=' ', file=sine_file)
        print(hx, end=' ')
    print()
    print(file=sine_file)  # add newline at the end of the file
```
